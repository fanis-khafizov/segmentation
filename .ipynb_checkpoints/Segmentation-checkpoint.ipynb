**4). Предобработка данных**

```python
TRAIN_DIR = Path(os.path.join(HOME_PATH, 'Segmentation/data/train'))
VAL_DIR = Path(os.path.join(HOME_PATH, 'Segmentation/data/val'))
TEST_DIR = Path(os.path.join(HOME_PATH, 'Segmentation/data/test'))
```

```python
import imageio
import numpy as np
import imgaug as ia
import imgaug.augmenters as iaa
from imgaug.augmentables.segmaps import SegmentationMapsOnImage
import torchvision.transforms as T

# преобразуем изображения и маски
def get_transforms(mode, image, mask):
    mask = SegmentationMapsOnImage(mask, shape=image.shape)
    
    aug = []
    # аугментации
    if mode != 'test':
        aug.append(iaa.CropToFixedSize(width=RESCALE_SIZE, height=RESCALE_SIZE))
    if mode == 'train':
        aug.append(iaa.Fliplr(0.5))
        aug.append(iaa.imgcorruptlike.GaussianNoise(severity=1))
        #aug.append(iaa.JpegCompression(compression=(65, 75)))
        aug.append(iaa.SaltAndPepper(0.005))
    
    aug = iaa.Sequential(aug)
    image, mask = aug(image=image, segmentation_maps=mask)
    image = np.array(image, dtype=np.float32) / 255
    image = torch.as_tensor(image, dtype=torch.float32)
    image = image.permute([2, 0, 1])
    image = T.Normalize([0.17583187, 0.18067491, 0.17738238], [0.28766868, 0.32577008, 0.28470638])(image)
    mask = mask.get_arr()
    mask = torch.as_tensor(mask, dtype=torch.uint8)
    return image, mask
```

```python
class SegDataset(torch.utils.data.Dataset):
    def __init__(self, root, transforms, mode):
        self.root = root
        self.mode = mode
        self.transforms = transforms
        self.imgs = list(sorted(os.listdir(os.path.join(root, 'images'))))
        self.masks = list(sorted([i for i in os.listdir(os.path.join(root, 'masks')) if 'labelIds.png' in i]))

    # функция для получения картинки и маски
    def __getitem__(self, idx):
        img_path = os.path.join(self.root, 'images', self.imgs[idx])
        mask_path = os.path.join(self.root, 'masks', self.masks[idx])
        img = Image.open(img_path).convert("RGB")
        img = np.array(img, dtype=np.uint8)
        mask = Image.open(mask_path)
        mask = np.array(mask, dtype=np.uint8)
        if self.transforms is not None:
            img, mask = self.transforms(self.mode, img, mask)
        img = torch.tensor(img, dtype=torch.float32)
        masks = mask == torch.arange(0, n_classes)[:, None, None]
        masks = torch.as_tensor(masks, dtype=torch.float32)
        return img, masks

    def __len__(self):
        return len(self.imgs)
    # функция, возвращающая картинку по индексу
    def get_img(self, idx):
        img_path = os.path.join(self.root, 'images', self.imgs[idx])
        img = Image.open(img_path).convert("RGB")
        img = np.array(img, dtype=np.uint8)
        return img
    
    def image_id(self, idx):
        return self.imgs[idx]
```

```python
train_dataset = SegDataset(TRAIN_DIR, get_transforms, 'train')
val_dataset = SegDataset(VAL_DIR, get_transforms, 'val')
test_dataset = SegDataset(VAL_DIR, get_transforms, 'test')
```

**5). Обучение модели для сегментации** 

```python
# объявляем функцию для fit'а
def fit_epoch(model, train_loader, criterion, optimizer):
    # обнуляем переменные
    running_loss = 0.0
    running_corrects = 0
    processed_data = 0

    # итерируемся по датасету
    for inputs, segmaps in tqdm(train_loader, desc='Fit'):
        if len(inputs) == 1:
            continue
        inputs = inputs.to(DEVICE)
        segmaps = segmaps.to(DEVICE)
        # обнуляем градиент
        optimizer.zero_grad()

        # делаем предсказания
        outputs = nn.Sigmoid()(model(inputs)['out'])
        # вычислем функцию потерь
        loss = criterion(outputs, segmaps)
        # распространяем ошибку на все предыдущие слои
        loss.backward()
        # делаем шаг оптимайзера
        optimizer.step()
        # определяем предссказанный класс как максимальный выход последнего слоя
        preds = torch.argmax(outputs, 1)
        preds = preds[:, None, :, :] == torch.arange(0, n_classes)[:, None, None].to(DEVICE)
        # прибавляем главные показатели за эпоху
        running_loss += loss.item() * inputs.size(0)
        running_corrects += torch.sum(preds == segmaps.data)
        processed_data += inputs.size(0)
    # считаем средний loss и accuracy
    train_loss = running_loss / processed_data
    train_acc = running_corrects.cpu().numpy() / processed_data / RESCALE_SIZE ** 2 / n_classes
    return train_loss, train_acc
```

```python
# функция для валидации
def eval_epoch(model, val_loader, criterion):
    # переводим модель в режим валидации
    model.eval()
    # обнуляем loss и другие величины
    running_loss = 0.0
    running_corrects = 0
    processed_size = 0

    # итерируемся по даталоадеру
    for inputs, segmaps in tqdm(val_loader, desc='Validation'):
        if len(inputs) == 1:
            continue
        # переносим данные на наше устройство
        inputs = inputs.to(DEVICE)
        segmaps = segmaps.to(DEVICE)

        # отключаем вычисление градиента
        with torch.set_grad_enabled(False):
            # делаем предсказание и вычисляем loss
            outputs = nn.Sigmoid()(model(inputs)['out'])
            loss = criterion(outputs, segmaps)
            preds = torch.argmax(outputs, 1)
            preds = preds[:, None, :, :] == torch.arange(0, n_classes)[:, None, None].to(DEVICE)

        # прибавляем loss, количество правилных предсказаний и общее число предсказаний за эпоху
        running_loss += loss.item() * inputs.size(0)
        running_corrects += torch.sum(preds == segmaps.data)
        processed_size += inputs.size(0)
    # вычисляем средний loss и accuracy
    val_loss = running_loss / processed_size
    val_acc = running_corrects.double() / processed_size / RESCALE_SIZE ** 2 / n_classes
    return val_loss, val_acc
```

```python
from torch.utils.data import DataLoader
import torch.nn as nn
from tqdm.notebook import tqdm, trange
# главная функция для обучения
def train(train_files, val_files, model, epochs, batch_size):
    # объявляем даталоадеры
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
    # объявляем numpy array, в который будем записывать loss и accuracy
    history = np.zeros((epochs, 4))

    # шаблон для вывода данных по ходу обучения
    log_template = "\nEpoch {ep:03d} train_loss: {t_loss:0.4f} \
    val_loss {v_loss:0.4f} train_acc {t_acc:0.4f} val_acc {v_acc:0.4f}"

    # объявляем оптимизатор и функцию потерь
    opt = torch.optim.Adam(model.parameters())
    criterion = nn.BCELoss()

    # делаем необходимое количество эпох
    for epoch in trange(epochs, desc='Total'):
        # производим fit
        train_loss, train_acc = fit_epoch(model, train_loader, criterion, opt)
        
        # делаем валидацию
        val_loss, val_acc = eval_epoch(model, val_loader, criterion)
        # записываем значение функции потерь и метрик
        history[epoch] = [train_loss, train_acc, val_loss, val_acc]
        # выводим значения за эпоху
        tqdm.write(log_template.format(ep=epoch+1, t_loss=train_loss,\
                                       v_loss=val_loss, t_acc=train_acc, v_acc=val_acc))
    # возвращаем loss и accuracy за эпохи
    return history
```

```python
model = torchvision.models.segmentation.deeplabv3_resnet50(pretrained=False, progress=True)
model.classifier[-1] = nn.Conv2d(256, n_classes, kernel_size=(1, 1), stride=(1, 1))
```

```python
model.load_state_dict(torch.load('DeepLabV3_resnet50.pth'))
```

```python
model = model.to(DEVICE)
```

```python
history = train(train_files=train_dataset, val_files=val_dataset, model=model, epochs=10, batch_size=2)
```

Для наглядности пропустим одну картинку полностью по частям

```python
def test(model, dataset, index):
    model.eval()
    height = dataset[0][0].shape[1]
    width = dataset[0][0].shape[2]
    preds = torch.zeros((height, width), dtype=torch.uint8)
    
    inputs = dataset[index][0]
    inputs = inputs.to(DEVICE)
    for x in range(0, height, RESCALE_SIZE):
        for y in range(0, width, 2 * RESCALE_SIZE):
            with torch.set_grad_enabled(False):
                preds[x:x + RESCALE_SIZE, y:y + 2 * RESCALE_SIZE] = \
                torch.argmax(model(inputs[None, :, x : x + RESCALE_SIZE, y : y + 2 * RESCALE_SIZE])['out'], 1).cpu()
    return preds
```

```python
import random
idx = random.randint(0, len(test_dataset))
test_preds = test(model, test_dataset, idx)
fig, axes = plt.subplots(nrows=2, figsize=(15, 17))
axes[0].imshow(test_dataset.get_img(idx))
axes[1].imshow(np.array(test_preds))
```

**6). Измерения качества работы модели (метрики согласуются с руководителем и зависят от задачи)**

```python
import sklearn.metrics
acc = np.zeros((len(test_dataset), n_classes), dtype=np.float64)
for i in tqdm(range(len(test_dataset))):
    outputs = test(model, test_dataset, i)
    outputs = np.array(outputs, dtype=np.uint8) == np.arange(0, n_classes)[:, None, None]
    target = np.array(test_dataset[i][1])
    for j in range(n_classes):
        intersection = np.logical_and(target[j], outputs[j])
        union = np.logical_or(target[j], outputs[j])
        acc[i][j] = np.sum(intersection) / np.sum(union)
```

```python
accuracy = np.mean(acc, axis=0)
```
